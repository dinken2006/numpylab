```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'London', 'Paris']
})

print(df['Name'])
print(df[['Name', 'Age']])

print(df.iloc[0])
print(df.iloc[0:2])
print(df.loc[1])

print(df[df['Age'] > 25])
print(df[df['City'] == 'London'])

df['Salary'] = [50000, 60000, 70000]
print(df)

df.loc[0, 'Age'] = 26
print(df)

print(df.sort_values('Age'))
print(df.sort_values('Salary', ascending=False))

print(df.describe())

df2 = pd.DataFrame({
    'Name': ['Alice', np.nan, 'Charlie'],
    'Age': [25, np.nan, 35]
})

print(df2.isnull())
print(df2.isnull().sum())

df2['Age'] = df2['Age'].fillna(df2['Age'].mean())
print(df2)

dept = pd.DataFrame({
    'Department': ['IT', 'HR', 'IT', 'HR'],
    'Salary': [50000, 40000, 60000, 45000]
})

print(dept.groupby('Department')['Salary'].mean())
```
